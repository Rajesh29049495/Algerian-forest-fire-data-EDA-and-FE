###### Check Multicollinearity for Categorical features
* A chi-squared test (also chi-square or χ2 test) is a statistical hypothesis test that is valid to perform when the test statistic is chi-squared distributed under the null hypothesis, specifically Pearson's chi-squared test

* A chi-square statistic is one way to show a relationship between two categorical variables.

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in categorical_features:
    if chi2_contingency(pd.crosstab(df['region'], df[feature]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result = pd.DataFrame(data=[categorical_features, chi2_test]).T
result.columns = ['Column', 'Hypothesis Result']
result
```

```python
###this above code would have done ch2 test on all the categorical columns of the dataframe, by applying ch2 test on each categorical column and 'region' categorical column.
#if we want to do the chi2 test on only categorical columns, as here in this dataset there are only two categorical columns
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['region'], df['Classes'])
print(contingency_table)
```

```python
##chi-square test of independece

c, p, dof, expected= chi2_contingency(contingency_table)
#where, c = chi2= the test statistic; p=p-value of test; dof=degree of freedom; expected= the expected frequencies, based on the marginal sums of the table
```

```python
p
```

```python
alpha = 0.05
```

```python
if p<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

###### Observation:
there is a relationship between 'Classes' and 'region' features of the dataset
